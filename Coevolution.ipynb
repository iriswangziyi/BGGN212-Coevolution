```python
# BGGN 212

import math
import random

# Input parameters ####################

# int; total population
X = 10
Y = 10

# float; maximum elapsed time
T = 100.0 

# float; start time
t = 0.0

# float; spatial parameter
V = 100.0 #TODO

# float; rate of ??
_alpha1 = 1.0
_beta1 = 0.5
_alpha2 = 1.0
_beta2 = 0.5

#########################################

# Initialize results list
data = []
data.append((t, X, Y))

# Main loop
while t < T:
    if X == 0 | Y==0:
        break

    # Species 1 makes a move
    w1 = _alpha1 * X
    w2 = _beta1 * X
    W = w1 + w2

    dt = -math.log(random.uniform(0.0, 1.0)) / W
    t = t + dt

    if random.uniform(0.0, 1.0) < w1 / W:
        #A1
        X = X + 1
        Y = Y - 1
    else:
        #S1
        X = X + 1
        Y = Y + 1

    data.append((t, X, Y))

    # Species 2 makes a move
    w1 = _alpha2 * Y
    w2 = _beta2 * Y
    W = w1 + w2

    dt = -math.log(random.uniform(0.0, 1.0)) / W
    t = t + dt

    if random.uniform(0.0, 1.0) < w1 / W:
        #A2
        X = X - 1
        Y = Y + 1
    else:
        #S2
        X = X + 1
        Y = Y + 1

    data.append((t, X, Y))
    
    
with open('Coevo_data.txt', 'w+') as fp:
    fp.write('\n'.join('%f %i %i' % x for x in data))
    
```
